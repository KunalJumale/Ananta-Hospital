```python
# Q. Create a new column age_group that labels patients as 'Child' (<18), 'Adult' (18–59), or 'Senior' (60+).
def age_group(age):
    if age < 18:
        return "Child"
    elif 18 <= age <= 59:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(age_group)
```

```python
# Q. Create a column total_dosage_mg = dosage * cost (for rows with cost & dosage).
df["total_dosage_mg"]=df["dosage_mg"]*df["cost"]
```
